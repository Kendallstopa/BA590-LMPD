```python
sf = shapefile.Reader("KentuckyZipCodes")  
 
shape_recs = sf.shapeRecords()
 
fig = plt.figure(figsize = (40,20))
#fig.set_size_inches(40.5, 20.5, forward=True)

ax = fig.add_subplot(111)
 
for rec in shape_recs:
    points = rec.shape.points
    d = rec.record  # record MetaData
    zip_code = d[2]
    c=(0.9, 0.2, 0.5, 0.2)

    
    if zip_code == "40212":  # Coloring in 1 zipcode
        c='blue'
    patch = patches.Polygon(points,True,label=d[2], color=c, alpha=0.09)
    ax.add_patch(patch)
    
    if zip_code == "40211":  # Coloring in 1 zipcode
        c='red'
    patch = patches.Polygon(points,True,label=d[2], color=c)
    ax.add_patch(patch)
    
    if zip_code == "40258":  # Coloring in 1 zipcode
        c='green'
    patch = patches.Polygon(points,True,label=d[2], color=c)
    ax.add_patch(patch)
    
    if zip_code == "40208":  # Coloring in 1 zipcode
        c='yellow'
    patch = patches.Polygon(points,True,label=d[2], color=c)
    ax.add_patch(patch)
    
    if zip_code == "40205":  # Coloring in 1 zipcode
        c='purple'
    patch = patches.Polygon(points,True,label=d[2], color=c)
    ax.add_patch(patch)
    
    if zip_code == "40228":  # Coloring in 1 zipcode
        c='black'
    patch = patches.Polygon(points,True,label=d[2], color=c)
    ax.add_patch(patch)
    
ax.autoscale()


plt.axis('off')


plt.show()
```
